```python
data.drop(["encounter_id" ,"patient_id" ,"hospital_id","Unnamed: 83", ],axis =1, inplace=True)
```

```python
Num=[]
for col in data.columns:
    if (data[col].dtype==int)or (data[col].dtype==float):
        Num.append(col)
print (Num)
```

```python
for col in data.columns:
    if col in Num :
        data[col].fillna(data[col].median(), inplace=True)
```

```python
categorical=[]
for col in data.columns:
    if  (data[col].dtype==object):
        categorical.append(col)
print (categorical)
```

```python
for col in data.columns:
    if col in categorical:
        data[col].fillna(data[col].mode(), inplace=True)
```

# PCA: Principal Component Analysis

1. > Principal Component Analysis, or PCA, is a dimensionality-reduction method that is often used to reduce the dimensionality of large data sets, by transforming a large set of variables into a smaller one that still contains most of the information in the large set.

![](https://i2.wp.com/dataaspirant.com/wp-content/uploads/2021/01/9-Principal-Component-Analysis-With-Dimensions.png?ssl=1)

1. > PCA is effected by scale so you need to scale the features in your data before applying PCA. Use RobustScaler to help you standardize the dataset’s features

```python
from category_encoders import CountEncoder
CE = CountEncoder(normalize=True, cols=categorical)
data = CE.fit_transform(data)
```

```python
from sklearn.preprocessing import RobustScaler
RS = RobustScaler()
scale = RS.fit_transform(data)
data = pd.DataFrame(scale, columns=data.columns)
```

```python
from sklearn.decomposition import PCA
pca = PCA()
pca.fit(data)
pca_data = pca.transform(data)
```

```python
data.shape
```

```python
component_names = [f"Component {i}" for i in range(1,82 )]
component_names
```

```python
pca_data = pd.DataFrame(pca_data, columns=component_names)
pca_data
```

# Explained Variance
1. > The explained variance tells you how much information (variance) can be attributed to each of the principal components. This is important as while you can convert 4 dimensional space to 2 dimensional space, you lose some of the variance (information) when you do this. By using the attribute explained_variance_ratio_,

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(20,10))
sns.barplot(x=list(range(1, 82)), y=pca.explained_variance_ratio_, palette="pastel")
```

# T-SNE: T-distributed Stochastic Neighbor Embedding

1. > t-SNE uses a heavy-tailed Student-t distribution with one degree of freedom to compute the similarity between two points in the low-dimensional space rather than a Gaussian distribution. T- distribution creates the probability distribution of points in lower dimensions space, and this helps reduce the crowding issue.

![](https://miro.medium.com/max/701/1*6j41Z2p5MW-YlzwBcjpG1Q.png)

```python
from sklearn.manifold import TSNE
tsne = TSNE(random_state=10)
tsne_data = tsne.fit_transform(data)
```

```python
tsne_data.shape
```

```python
plt.figure(figsize=(20,12))
sns.scatterplot(x=tsne_data[:,0], y=tsne_data[:,1] ,hue=data["hospital_death"], palette='pastel')
```

# UMAP: Uniform Manifold Approximation and Projection for Dimension Reduction

1. > Uniform Manifold Approximation and Projection (UMAP) is a dimension reduction technique that can be used for visualisation similarly to t-SNE, but also for general non-linear dimension reduction. The algorithm is founded on three assumptions about the data

```python
import umap
ump = umap.UMAP(random_state=10)
umap_data = ump.fit_transform(data)
```

```python
plt.figure(figsize=(20,12))
sns.scatterplot(x=umap_data[:,0], y=umap_data[:,1], hue=data["hospital_death"], palette='pastel')
plt.show()
```
